**Conditional Statement**

```python
# Contoh singkat penggunaan operator relasional di Python
# Mendefinisikan dua angka
angka1 = 10
angka2 = 5

# Operator '==' : Memeriksa kesamaan
sama = angka1 == angka2
# Operator '!=' : Memeriksa ketidaksamaan
tidak_sama = angka1 != angka2
# Operator '>' : Memeriksa lebih besar
lebih_besar = angka1 > angka2
# Operator '<' : Memeriksa lebih kecil
lebih_kecil = angka1 < angka2
# Operator '>=' : Memeriksa apakah lebih besar atau sama dengan
lebih_besar_sama_dengan = angka1 >= angka2
# Operator '<=' : Memeriksa apakah lebih kecil atau sama dengan
lebih_kecil_sama_dengan = angka1 <= angka2

# Cetak hasilnya
print("Apakah sama?", sama)
print("Apakah tidak sama?", tidak_sama)
print("Apakah lebih besar?", lebih_besar)
print("Apakah lebih kecil?", lebih_kecil)
print("Apakah lebih besar sama dengan?", lebih_besar_sama_dengan)
print("Apakah lebih kecil sama dengan?", lebih_kecil_sama_dengan)

# output code
# Apakah sama? False
# Apakah tidak sama? True
# Apakah lebih besar? True
# Apakah lebih kecil? False
# Apakah lebih besar atau sama dengan? True
# Apakah lebih kecil atau sama dengan? False
```

**Latihan - Conditional Statement**

Buatlah script yang memberikan output sebagai
berikut:

*   Jika angka kelipatan 3, beri output “fizz”

*   Jika angka kelipatan 5, beri output “buzz”

*   Jika angka kelipatan 3 dan 5 beri output “fizzbuzz”

*   Selain itu beri output “Nayy”

```python
# Fungsi untuk cek kondisi angka
def cek_fizzbuzz(angka):
    if angka % 3 == 0 and angka % 5 == 0:
        return "fizzbuzz"
    elif angka % 3 == 0:
        return "fizz"
    elif angka % 5 == 0:
        return "buzz"
    else:
        return "Nayy"

# Contoh penggunaan
for i in range(1, 21):
    print(f"Angka {i}: {cek_fizzbuzz(i)}")
```

**LOOP PYTHON**

```python
# Contoh singkat penggunan pernyataan while-loop di Pyhton

# Inisiasi sebuah variabel
hitung = 1

# Pernyataan whila : Mnegeksekusi blok kode selama kondisi terpenuhi
while hitung <= 5:
  print(f"Perulangan ke-{hitung}")
  hitung += 1 # Menginkrementasi variabel hitung
```

```python
# List hari dalam seminggu
hari_minggu = ["Senin", "Selasa", "Rabu", "Kamis", "Jumat", "Sabtu", "Minggu"]

# Inisialisasi variabel
index = 6

# Perulangan while
while index < len(hari_minggu):
    print(hari_minggu[index])
    index += 1
```

```python
# Contoh singkat penggunaan pernyataan for di python

# Pernyataan for : Iterasi melalui elemen dalam sebuah list
buah = ["apel", "pisang", "jeruk"]

# Mengeksekusi blok kode untuk setiap elemen dalam list
for buah_sekarang in buah:
  print(f"Sekarang ada {buah_sekarang}")
```

```python
# Contoh singkat penguunaan nested loop di Python '

# Pernytaan nested liip: Iterasi melalui baris dan kolom
for baris in range (1, 4):
  for kolom in range (1, 4):
    print(f"Baris {baris}, Kolom {kolom}")
```
