**Cek duplikat**

```python
data.duplicated()
```

```python
data['duplicated']=data.duplicated()
data
```

handle duplicated

```python
data_new = data.drop_duplicates()
data_new
```

**Filter dataframe**

```python
data_more_20000 = data[data['price'] > 20000]
data_more_20000
```

multiple filter

```python
multiple_filter = data[(data['price'] > 20000) & (data['year'] == 2015)]
multiple_filter
```

```python
data.query('price > 20000 and year == 2015')
```

**Grouping**

```python
data.groupby('year')['price'].mean()
```

```python
data.groupby('year')['price'].agg(['mean', 'max', 'min'])
```
