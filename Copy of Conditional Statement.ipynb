# Conditional Statement

## Operator Perbandingan

```python
#@title Kesamaan

angka1 = 10
angka2 = 5

sama = angka1 == angka2

print("Apakah sama ?", sama)
```

```python
#@title Tidak Sama

angka1 = 10
angka2 = 5

tidak_sama = angka1 != angka2

print("Apakah tidak sama ?", tidak_sama)
```

```python
#@title Lebih Besar

angka1 = 10
angka2 = 5

lebih_besar = angka1 > angka2

print("Apakah lebih besar ?", lebih_besar)
```

```python
#@title Lebih Kecil

angka1 = 10
angka2 = 5

lebih_kecil = angka1 < angka2

print("Apakah lebih kecil ?", lebih_kecil)
```

```python
#@title Lebih Besar Sama Dengan

angka1 = 10
angka2 = 5

lebih_besar_sama_dengan = angka1 >= angka2

print("Apakah lebih besar atau sama dengan ?", lebih_besar_sama_dengan)
```

```python
#@title Lebih Kecil Sama Dengan

angka1 = 10
angka2 = 5

lebih_kecil_sama_dengan = angka1 <= angka2

print("Apakah lebih besar atau sama dengan ?", lebih_kecil_sama_dengan)
```

## Operator Logika

```python
#@title AND

kondisi1 = True
kondisi2 = False

logika_and = kondisi1 and kondisi2

print("Hasil Logika And:", logika_and)
```

```python
#@title OR

kondisi1 = False
kondisi2 = False

logika_or = kondisi1 or kondisi2

print("Hasil Logika OR:", logika_or)
```

```python
#@title NOT

kondisi1 = True
kondisi2 = False

logika_not_1 = not kondisi1
logika_not_2 = not kondisi2

print("Hasil Logika NOT Kondisi 1:", logika_not_1)
print("Hasil Logika NOT Kondisi 2:", logika_not_2)
```

## Condition IF

```python
#@title if

bilangan = 6

if bilangan > 4:
    print("Bilangan positif")
```

```python
#@title if-else

bilangan = 5

if bilangan > 0:
    print("Bilangan positif")
else:
    print("Bilangan negatif")
```

```python
#@title if-elif-else

bilangan = 0

if bilangan > 0:
    print("Bilangan positif")
elif bilangan < 0:
    print("Bilangan negatif")
else:
    print("Nol")
```

## Soal 2

Buatlah script yang memberikan output sebagai berikut:
Jika angka kelipatan 3, beri output “fizz”
Jika angka kelipatan 5, beri output “buzz”
Jika angka kelipatan 3 dan 5 beri output “fizzbuzz”
Selain itu beri output “Nayy”


```python
# prompt: Buatlah script yang memberikan output sebagai berikut: Jika angka kelipatan 3, beri output “fizz” Jika angka kelipatan 5, beri output “buzz” Jika angka kelipatan 3 dan 5 beri output “fizzbuzz” Selain itu beri output “Nayy

angka = int(input("Masukkan angka: "))

if angka % 3 == 0 and angka % 5 == 0:
  print("fizzbuzz")
elif angka % 3 == 0:
  print("fizz")
elif angka % 5 == 0:
  print("buzz")
else:
  print("Nayy")
```

## Soal 3

Buat contoh dari :
* Operator perbandingan
* Operator Logika
* Operator Condition

```python
# prompt: Buat contoh dari :
# Operator perbandingan
# Operator Logika
# Operator Condition

# Operator Perbandingan
angka1 = 15
angka2 = 10

print("Apakah angka1 sama dengan angka2?", angka1 == angka2)
print("Apakah angka1 lebih besar dari angka2?", angka1 > angka2)
print("Apakah angka1 lebih kecil dari atau sama dengan angka2?", angka1 <= angka2)


# Operator Logika
kondisi1 = True
kondisi2 = False

print("Hasil AND:", kondisi1 and kondisi2)
print("Hasil OR:", kondisi1 or kondisi2)
print("Hasil NOT kondisi1:", not kondisi1)


# Operator Condition (If-Elif-Else)
nilai = 75

if nilai >= 80:
  print("Nilai Anda A")
elif nilai >= 70:
  print("Nilai Anda B")
elif nilai >= 60:
  print("Nilai Anda C")
else:
  print("Nilai Anda D")
```

## Soal 4

buah = ["nanas", "apel", "jeruk"]

buat conditional if dengan loop

```python
# prompt: buah = ["nanas", "apel", "jeruk"]
# buat conditional if dengan loop

buah = ["nanas", "apel", "jeruk"]

for item in buah:
  if item == "apel":
    print(f"{item} adalah buah favoritku!")
  else:
    print(f"{item} juga enak.")
```
